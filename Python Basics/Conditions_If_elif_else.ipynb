```python
#CONDITIONS
```

```python
#Simple if
age = 18

if age >= 18:
    print("You can vote")
```

```python
#if-else
num = 10

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#if-elif-else
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

```python
#Multiple Conditions (and)
age = 20
has_id = True

if age >= 18 and has_id:
    print("Allowed")
else:
    print("Not Allowed")
```

```python
#Multiple Conditions (or)
day = "Sunday"

if day == "Sunday" or day == "Saturday":
    print("Holiday")
```

```python
#not operator
is_logged_in = False

if not is_logged_in:
    print("Please login")
```

```python
#Nested if
num = 15

if num > 0:
    if num % 5 == 0:
        print("Positive & divisible by 5")
```

```python
#Ternary (Short if-else)
num = 7

result = "Even" if num % 2 == 0 else "Odd"
print(result)
```

```python
#Check in list
fruits = ["apple", "banana"]

if "apple" in fruits:
    print("Found")
```

```python
#Check in dictionary
student = {"name": "Ankit", "age": 21}

if "name" in student:
    print("Key exists")
```

```python
#Input Based Condition
num = int(input("Enter number: "))

if num > 0:
    print("Positive")
elif num == 0:
    print("Zero")
else:
    print("Negative")
```
