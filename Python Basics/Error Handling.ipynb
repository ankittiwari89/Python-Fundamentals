```python
#Error handdling
```

```python
#Simple try-except
try:
    x = int("abc")   # error
except:
    print("Error occurred")
```

```python
#Specific Exception
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
#Multiple Exceptions
try:
    num = int("abc")
    result = 10 / num
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Division by zero")
```

```python
#else block
try:
    num = int("10")
except ValueError:
    print("Error")
else:
    print("Success:", num)
```

```python
#finally block
try:
    print("Try block")
except:
    print("Error")
finally:
    print("Always runs")
```

```python
#Exception as variable
try:
    x = 10 / 0
except Exception as e:
    print("Error:", e)
```

```python
#Raise Exception
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

```python
#Custom Exception Message
try:
    num = int(input("Enter number: "))
    if num < 0:
        raise Exception("Negative not allowed")
except Exception as e:
    print(e)
```

```python
#try-except in function
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

print(divide(10, 0))
```

```python
#File Handling with Exception
try:
    with open("data.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found")
```

```python
#Multiple errors in one except
try:
    x = int("abc")
except (ValueError, TypeError):
    print("Error handled")
```

```python
#pass in except
try:
    x = int("abc")
except:
    pass   # ignore error
```
