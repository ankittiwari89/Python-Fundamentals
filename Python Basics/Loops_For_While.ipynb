```python
#For LOOPS
```

```python
#Simple for loop
for i in range(5):
    print(i)
```

```python
#range(start, stop)
for i in range(1, 6):
    print(i)
```

```python
#range(step)
for i in range(1, 10, 2):
    print(i)
```

```python
#Loop through list
nums = [1, 2, 3, 4]

for i in nums:
    print(i)
```

```python
#Loop through string
text = "Python"

for ch in text:
    print(ch)
```

```python
#enumerate()
fruits = ["apple", "banana"]

for index, value in enumerate(fruits):
    print(index, value)
```

```python
#break
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
#continue
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
#pass
for i in range(3):
    pass
```

```python
#Nested for loop
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#WHILE LOOP
```

```python
#Simple while
i = 0

while i < 5:
    print(i)
    i += 1
```

```python
#break in while
i = 0

while True:
    if i == 3:
        break
    print(i)
    i += 1
```

```python
#continue in while
i = 0

while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
#while with else
i = 0

while i < 3:
    print(i)
    i += 1
else:
    print("Done")
```

```python
#Input based while loop
while True:
    num = int(input("Enter number (0 to stop): "))
    
    if num == 0:
        break
    
    print("You entered:", num)
```
