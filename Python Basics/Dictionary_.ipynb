```python
#Create Dictionary
student = {
    "name": "Ankit",
    "age": 21,
    "course": "Python"
}

print(student)
```

```python
#Access Values
print(student["name"])      
print(student.get("age"))  
```

```python
#Add / Update Value
student["city"] = "Delhi"   # add
student["age"] = 22         # update

print(student)
```

```python
#Keys(), values(), items()
print(student.keys())
print(student.values())
print(student.items())
```

```python
#Pop()
student.pop("course")   # remove by key
print(student)
```

```python
#popitem()
student.popitem()   # last inserted remove
print(student)
```

```python
#clear()
temp = {"a": 1, "b": 2}
temp.clear()
print(temp)
```

```python
#update()
student.update({"age": 25, "country": "India"})
print(student)
```

```python
#copy()
new_student = student.copy()
print(new_student)
```

```python
#setdefault()
student.setdefault("state", "UP")   # add if not exist
print(student)
```

```python
#fromkeys()
keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)

print(new_dict)
```

```python
#len() & type()
print(len(student))
print(type(student))
```

```python
#Loop Dictionary
for key, value in student.items():
    print(key, ":", value)
```

```python
#Check Key Exists
print("name" in student)
print("salary" in student)
```
