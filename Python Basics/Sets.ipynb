```python
#Create Set
my_set = {1, 2, 3, 2, 4}
print(my_set)  # duplicates removed
```

```python
#Basic Info
print(len(my_set))
print(type(my_set))
```

```python
#Add() function
my_set.add(5)
print(my_set)
```

```python
#Remove() function
my_set.remove(1)
print(my_set)
```

```python
#Discard() function
my_set.discard(10)   # no error if not present
print(my_set)
```

```python
#Pop() function
my_set.pop()   # random element remove
print(my_set)
```

```python
#Clear() function
temp_set = {1, 2, 3}
temp_set.clear()
print(temp_set)
```

```python
#SET OPERATIONS
```

```python
#Union
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.union(set2))
```

```python
#Intersection
print(set1.intersection(set2))
```

```python
#Difference
print(set1.difference(set2))
```

```python
#Symmetric Difference
print(set1.symmetric_difference(set2))
```
