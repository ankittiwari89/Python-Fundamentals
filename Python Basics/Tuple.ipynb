```python
#Create Tuple
my_tuple = (1, 2, 3, 2, 4)
print(my_tuple)
```

```python
#Basic Info
print(len(my_tuple))     # length
print(type(my_tuple))    # type
```

```python
# Access Element
print(my_tuple[0])   # first element
print(my_tuple[-1])  # last element
```

```python
#Count() Function
print(my_tuple.count(2))   # count occurrences
```

```python
# Index() Function
print(my_tuple.index(3))   # index of element
```

```python
#Loop
for i in my_tuple:
    print(i)
```

```python
#Tuple Packing & Unpacking
a, b, c = (10, 20, 30)
print(a, b, c)
```

```python
#Convert Tuple to List
my_tuple = (1, 2, 3, 4)
my_list = list(my_tuple)

print(my_list)
```

```python
#Append()
my_list.append(5)
print(my_list)
```

```python
#Insert()
my_list.insert(1, 10)   # index, value
print(my_list)
```

```python
#Extend()
my_list.extend([6, 7, 8])
print(my_list)
```

```python
#Remove()
my_list.remove(2)   # value remove
print(my_list)
```

```python
#Pop()
my_list.pop()       # last remove
print(my_list)

my_list.pop(1)      # index wise remove
print(my_list)
```

```python
#Clear()
temp = [1, 2, 3]
temp.clear()
print(temp)
```

```python
#Update value
my_list[0] = 100
print(my_list)
```

```python
#Sort()
nums = [5, 2, 9, 1]

nums.sort()
print(nums)
```

```python
#Reverse()
nums.reverse()
print(nums)
```

```python
#Copy()
new_list = nums.copy()
print(new_list)
```

```python
#Count()
print(nums.count(2))
```

```python
#Index()
print(nums.index(5))
```

```python
#Len()
print(len(nums))
```

```python
#Convert Back List to Tuple
new_tuple = tuple(my_list)
print(new_tuple)
```
