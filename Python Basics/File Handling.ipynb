```python
#File Handling
```

```python
#Create File + Write (w)
with open("data.txt", "w") as file:
    file.write("Hello Ankit\n")
    file.write("Welcome to Python")
```

```python
#Read File (r)
with open("data.txt", "r") as file:
    print(file.read())
```

```python
#Read Line by Line
with open("data.txt", "r") as file:
    print(file.readline())   # first line
```

```python
#Read All Lines as List
with open("data.txt", "r") as file:
    print(file.readlines())
```

```python
#Append Data (a)
with open("data.txt", "a") as file:
    file.write("\nNew Line Added")
```

```python
#Write Multiple Lines
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]

with open("data.txt", "w") as file:
    file.writelines(lines)
```

```python
#Check File Exists
import os

print(os.path.exists("data.txt"))
```

```python
#Read + Write Together (r+)
with open("data.txt", "r+") as file:
    print(file.read())
    file.write("\nAdded using r+")
```

```python
#Append + Read (a+)
with open("data.txt", "a+") as file:
    file.write("\nAppend + Read")
    file.seek(0)
    print(file.read())
```

```python
#File Pointer (seek)
with open("data.txt", "r") as file:
    file.seek(5)
    print(file.read())
```

```python
#Tell Position
with open("data.txt", "r") as file:
    print(file.tell())
    file.read(5)
    print(file.tell())
```

```python
#Rename File
import os

os.rename("data.txt", "new_data.txt")
```

```python
#Delete File
import os

os.remove("new_data.txt")
```

```python
#Create File using x
with open("test.txt", "x") as file:
    file.write("New File Created")
```
