```python
# String 
name = "Rahul"
city = 'Delhi'
intro = "Main Data Science seekh raha hoon"

print(name)
print(city)
print(intro)
```

```python
# Strings (concatenation)
first_name = "Rahul"
last_name = "Sharma"

full_name = first_name + " " + last_name
print(full_name)   # Rahul Sharma

# f-string 
age = 25
intro = f"Mera naam {first_name} hai aur umar {age} saal hai"
print(intro)
```

```python
name = "Rahul Sharma"

print(name.upper())    # RAHUL SHARMA —  capital
print(name.lower())    # rahul sharma —  small
print(len(name))       # Length
```

```python
# strip =  extra spaces remove
name = "   Rahul   "

print(name.strip())    # "Rahul" — Remove space from both sides
print(name.lstrip())   # "Rahul   " — Remove space from left side
print(name.rstrip())   # "   Rahul" — Remove space from Right side
```

```python
# replace words
city = "Main Delhi mein rehta hoon"

new_city = city.replace("Delhi", "Noida")
print(new_city)   
```

```python
# split = Split string into parts 
name = "Rahul Sharma Delhi"

parts = name.split()
print(parts)       # ['Rahul', 'Sharma', 'Delhi']

# Split from comma
skills = "Python,Excel,Tally,GST"
skill_list = skills.split(",")
print(skill_list)  # ['Python', 'Excel', 'Tally', 'GST']
```

```python
# find word position
sentence = "Main Data Science seekh raha hoon"

print(sentence.find("Data"))     # 5 — position number
print(sentence.find("Python"))  

# count 
print(sentence.count("a"))       
```

```python
# startswith 
# endswith 
name = "Rahul Sharma"

print(name.startswith("Rahul"))   # True
print(name.startswith("Sharma"))  # False
print(name.endswith("Sharma"))    # True
print(name.endswith("Rahul"))     # False
```

```python
# Slicing 
# Format: string[start:end]
# Index start with 0
name = "Rahul Sharma"
#        0123456789...

print(name[0])      # R — Fisrt character
print(name[0:5])    # Rahul — 0 to 4 
print(name[6:])     # Sharma — 6 to end
print(name[:5])     # Rahul — start to 4
print(name[-1])     # a — Last Character
print(name[::-1])   # Reverse print
```

```python
# simple profile card
name = "  rahul sharma  "
city = "Delhi"
course = "Data Science"
skills = "Python,Excel,Tally"

name = name.strip()           # Remove Space
name = name.title()           # Proper Case 
skill_list = skills.split(",") # Split skills

print("Name:", name)
print("City:", city.upper())
print("Course:", course)
print("Skills:", skill_list)
print("Name starts with R?", name.startswith("R"))
print("First name:", name[0:5])
```
