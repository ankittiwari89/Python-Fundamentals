```python
# Arithmetic Operators
a = 10
b = 3

print(a + b)   # Addition = 13
print(a - b)   # Subtraction = 7
print(a * b)   # Multiplication = 30
print(a / b)   # Division = 3.33
print(a // b)  # Floor Division = 3 
print(a % b)   # Modulus = 1 (remainder)
print(a ** b)  # Power = 1000 (10  3rd power)
```

```python
# Comparison Operators 
# True False
a = 10
b = 3

print(a == b)   # Equal to = False
print(a != b)   # Not equal = True
print(a > b)    # Greater than = True
print(a < b)    # Less than = False
print(a >= b)   # Greater or equal = True
print(a <= b)   # Less or equal = False
```

```python
# Assignment Operator
a = 10        # simple assign
print(a)      # 10

a += 5        # a = a + 5
print(a)      # 15

a -= 3        # a = a - 3
print(a)      # 12

a *= 2        # a = a * 2
print(a)      # 24

a /= 4        # a = a / 4
print(a)      # 6.0
```

```python
# Logical Operators 
a = True
b = False

print(a and b)   # AND 
print(a or b)    # OR 
print(not a)     # NOT 

# Practical example
age = 25
salary = 20000

print(age > 18 and salary > 15000)  # True
print(age > 30 or salary > 15000)   # True
```

```python
# Identity Operators 
a = 10
b = 10
c = 20

print(a is b)      # True — same value
print(a is c)      # False — different value
print(a is not c)  # True — different 
```

```python
# Membership Operators
fruits = ["apple", "banana", "mango"]

print("apple" in fruits)      # True
print("grapes" in fruits)     # False
print("grapes" not in fruits) # True
```

```python
#salary calculator
salary = 20000
bonus = 5000
tax_rate = 0.10

total = salary + bonus           # Arithmetic
tax = total * tax_rate           # Arithmetic
final_salary = total - tax       # Arithmetic

print("Total:", total)
print("Tax:", tax)
print("Final Salary:", final_salary)

print("Salary above 20000?", final_salary > 20000)   # Comparison
print("Tax above 1000?", tax > 1000)                  # Comparison
```
