```python
#FUNCTIONS
```

```python
#Simple Function
def greet():
    print("Hello World")

greet()
```

```python
#Function with Parameters
def greet(name):
    print("Hello", name)

greet("Ankit")
```

```python
#Multiple Parameters
def add(a, b):
    print(a + b)

add(5, 3)
```

```python
#Function with Return
def square(num):
    return num * num

result = square(4)
print(result)
```

```python
#Default Parameters
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Ankit")
```

```python
#Keyword Arguments
def student(name, age):
    print(name, age)

student(age=21, name="Ankit")
```

```python
#Arbitrary Arguments (*args)
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)
```

```python
#Arbitrary Keyword Arguments (**kwargs)
def info(**data):
    print(data)

info(name="Ankit", age=21)
```

```python
#Return Multiple Values
def calc(a, b):
    return a + b, a - b

x, y = calc(10, 5)
print(x, y)
```

```python
#Lambda Function
square = lambda x: x * x
print(square(5))
```

```python
#Function inside Function
def outer():
    def inner():
        print("Inner function")
    inner()

outer()
```

```python
#Recursive Function
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
#Function with List
def show_list(lst):
    for i in lst:
        print(i)

show_list([1, 2, 3])
```

```python
#docstring (function description)
def add(a, b):
    """This function adds two numbers"""
    return a + b

print(add(2, 3))
```
