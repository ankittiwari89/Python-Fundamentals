```python
# List 
# Square brackets [] 
fruits = ["apple", "banana", "mango"]
numbers = [10, 20, 30, 40, 50]
mixed = ["Rahul", 25, True, 15000.50]

print(fruits)
print(numbers)
print(mixed)
```

```python
# Index start from 0
# -1 Last Item
fruits = ["apple", "banana", "mango", "grapes"]

print(fruits[0])    # apple — First item
print(fruits[1])    # banana — 2nd item
print(fruits[-1])   # grapes — last item
print(fruits[-2])   # mango — second last
```

```python
# Append = Add item end of the List
skills = ["Python", "Excel", "Tally"]
print(skills)   # ['Python', 'Excel', 'Tally']

skills.append("SQL")
print(skills)   # ['Python', 'Excel', 'Tally', 'SQL']

skills.append("Power BI")
print(skills)   # ['Python', 'Excel', 'Tally', 'SQL', 'Power BI']
```

```python
# insert= insert item any place
# insert(index, value)
skills = ["Python", "Excel", "Tally"]

skills.insert(1, "SQL")   
print(skills)   # ['Python', 'SQL', 'Excel', 'Tally']
```

```python
# remove = Remove item from Name
skills = ["Python", "Excel", "Tally", "SQL"]

skills.remove("Tally")
print(skills)   # ['Python', 'Excel', 'SQL']

# pop = (Remove last item by default)
skills.pop()
print(skills)   # ['Python', 'Excel']

skills.pop(0)   # Remove First item
print(skills)   # ['Excel']
```

```python
#Length,Count,Index
fruits = ["apple", "banana", "mango", "apple", "grapes"]

print(len(fruits))           # 5 — kitne items hain
print(fruits.count("apple")) # 2 — apple kitni baar hai
print(fruits.index("mango")) # 2 — mango kis index pe hai
```

```python
#Sort & Reverse
numbers = [50, 10, 40, 20, 30]

numbers.sort()
print(numbers)    # [10, 20, 30, 40, 50] — Ascending Order

numbers.reverse()
print(numbers)    # [50, 40, 30, 20, 10] — Descending Order
# Sort reverse order 
numbers.sort(reverse=True)
print(numbers)    # [50, 40, 30, 20, 10]
```

```python
# extend List
skills1 = ["Python", "Excel"]
skills2 = ["SQL", "Power BI"]

skills1.extend(skills2)
print(skills1)   # ['Python', 'Excel', 'SQL', 'Power BI']
```

```python
#Copy & Clear
skills = ["Python", "Excel", "SQL"]

# copy List
new_skills = skills.copy()
print(new_skills)   # ['Python', 'Excel', 'SQL']

# clear 
skills.clear()
print(skills)       # [] — empty list
```

```python
#Slicing
numbers = [10, 20, 30, 40, 50]

print(numbers[0:3])   # [10, 20, 30] — First 3
print(numbers[2:])    # [30, 40, 50] — 2 to end tak
print(numbers[:3])    # [10, 20, 30] — start to 3 
print(numbers[::-1])  # [50, 40, 30, 20, 10] — Reverse
```

```python
# In operators check
skills = ["Python", "Excel", "SQL"]

print("Python" in skills)    # True
print("Tally" in skills)     # False
print("Tally" not in skills) # True
```
